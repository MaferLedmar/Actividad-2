```python
#Carga desde un archivo .csv sin indice
data= pd.read_csv('inmuebles2.csv') 
data
```

```python
#Eliminar primera fila o filas innecesarias
data1= pd.read_csv('inmuebles2.csv', skiprows=1, low_memory=False)
data1
```

```python
#Eliminar columnas innecesarias
data2= data1.drop(['0','Referencia'], axis=1)
data2
```

```python
#Identificar valores nulos
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Realizamos una copia del dataframe
data3=data2.copy()
```

*Vemos que todas nuestras columnas tienen valores nulos, por lo que hay que realizar métodos adecuados para cada tipo de dato*

```python
#Sustituir valores nulos con promedio o media
data3["Precio Venta"]=data3["Precio Venta"].fillna(round(data2["Precio Venta"].mean(),1))
data3
```

```python
#Corroboramos valores nulos
valores_nulos=data3.isnull().sum()
valores_nulos
```

```python
#Sustituir valores nulos con mediana
data3["Superficie"]=data3["Superficie"].fillna(round(data2["Superficie"].median(),1))
data3
```

```python
#Corroboramos valores nulos
valores_nulos=data3.isnull().sum()
valores_nulos
```

```python
#Sustituir valores nulos por valores no nulos hacia adelante  "forward fill" ("ffill")
#Filtro por columnas

#Actividad_total=Alumnos.iloc[ : , [6]]
#Actividad_total.fillna(method="ffill")

data3['Fecha Alta'] = data3['Fecha Alta'].fillna(method="ffill")
```

```python
#Corroboramos valores nulos
valores_nulos=data3.isnull().sum()
valores_nulos
```

```python
#Sustituir valores nulos por un string en  concreto 
#Filtro por columnas

#Actividad_total=Alumnos.iloc[ : , [6]]

data3['Operación'] = data3['Operación'].fillna("Alquiler")  

#Corroboramos valores nulos
valores_nulos=data3.isnull().sum()
valores_nulos
```

```python
#Sustituir valores nulos por valores no nulos hacia atrás backward fill" ("bfill")
#Filtro por columnas

data3['Provincia'] = data3['Provincia'].fillna(method="bfill")

#Corroboramos valores nulos
valores_nulos=data3.isnull().sum()
valores_nulos
```

```python
#Sustituir valores nulos por valores no nulos hacia adelante  "forward fill" ("ffill")
#Filtro por columnas
#Sustitución de TIPO
data3['Tipo'] = data3['Tipo'].fillna(method="ffill")

#Corroboramos valores nulos
valores_nulos=data3.isnull().sum()
valores_nulos
```

```python
#Eliminamos filas con nulos
data4= data3.dropna(0)
data4
```

```python
#Corroboramos valores nulos
valores_nulos=data4.isnull().sum()
valores_nulos
```

```python
#Obtenemos la cuenta de los elementos nulos del dataframe completo
Cuenta_Total_Nulos= data4.isnull().sum().sum()
Cuenta_Total_Nulos
```

```python
#Realizamos histograma de variables cuantitativas
plt.hist(x=data4["Precio Venta"], color='blue', rwidth=0.50)
plt.title('Histograma de Precio de Venta')
plt.xlabel('Precio Venta')
plt.ylabel('Frecuencia')

plt.show() #dibujamos el histograma
```

```python
#Realizamos histograma de variables cuantitativas
plt.hist(x=data4["Superficie"], color='blue', rwidth=0.50)
plt.title('Histograma de Superficies')
plt.xlabel('Superficies')
plt.ylabel('Frecuencia')

plt.show() #dibujamos el histograma
```
